```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# create sample data X
X = np.random.randint(1, 30, (50, 1))
# create sample data y
y = 3.7 * X + np.random.randint(5, 30, (50, 1))

plt.scatter(X, y)
plt.show()

# now, we can express y in terms of x: y = Ax + B
model = LinearRegression()
model.fit(X, y)

print(model.coef_) # gradient
print(model.intercept_) # intercept

plt.scatter(X, y)
plt.plot(X, model.predict(X), "g")
plt.show()
```
