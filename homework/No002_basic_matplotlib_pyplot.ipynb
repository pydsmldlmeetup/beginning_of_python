```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, "o")
plt.show()
```

```python
a = np.array(range(100))
b = a ** 2

plt.plot(a, b)
plt.show()
```

```python
# you can show graph title ...
plt.plot(a, b)
plt.title("quadratic function")
plt.show()
```

```python
# you can also show x and y label ...
plt.plot(x, y)
plt.title("linear function")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# if you want to show two graph at one time...
plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.title("linear function")
plt.show()

plt.subplot(2, 1, 2)
plt.plot(a, b)
plt.title("quadratic function")
plt.show()
```

```python
# of course you can also show two data on the same graph...
x = np.array(range(100))
y = 70 * x + 1000

plt.plot(x, y, label="linear function")
plt.plot(a, b, label="quadratic function")

plt.legend(loc="lower right")
plt.show()
```
